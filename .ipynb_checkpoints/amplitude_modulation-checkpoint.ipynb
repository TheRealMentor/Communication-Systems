# Amplitude Modulation

We will write a code that will help us understand amplitude modulation in time and frequency domain. For this purpose, we will use python because it has inbuilt functions ans libraries for our help. We could have also done in **MATLAB**, but it is totally upto you to use in which language you code. Alright, let's get started.

To start our code, we need to begin with importing the necessary libraries that we will need in the program. We will import **numpy** for mathematical calculations and **matplotlib** for plotting graphs.

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

Now, after importing the necessary libraries, we will define the parameters to generate a carrier and message/modulation signal. A little background on carrier and message signals. We define them as follows:

$$c(t) = A_{c}cos(2\pi f_{c}t)...............(1)$$ 
$$m(t) = A_{m}cos(2\pi f_{m}t)...............(2)$$

$c(t)$ represents carrier signal and $m(t)$ represents message signal. For generating them, we need the amplitude and frequency of the wave. For our convenience, we take $A_{m} = 1$

```python
#Defining given parameters
A_c = 1      #Carrier Amplitude
f_c = 0.4    #Carrier Freq.
f_m = 0.05   #Modulation Freq.

#Defining time periods
T_c = 1/f_c  #Time period of carrier signal
T_m = 1/f_m  #Time period of modulating signal
```

### Function creating Modulated Signal

In this next block of code, we defined the range on the x-axis (time-axis) and we look at the $10$ cycles of the modulated wave, we is in another words $200$ sec on time axis. To get a smoother curve, we use $0.01$ step points. After defining the time-axis, we define the message/modulating signal and carrier signal. Using the defined signals, we generate a modulated signal, whose function I have defined below for the case of **single-tone modulation**.

```python
t = np.arange(0, 200, 0.1)             #Defining the x-axis range

message = np.cos(2*np.pi*f_m*t)           #Defining message signal

#Defining function to create modulated signal
def modulate(message, A_c, f_c, n_u):
    c = A_c*np.cos(2*np.pi*f_c*t)

    mod = (1 + n_u*message)*c
    return mod
```

### Modulated Signals

We, then, use the *modulate* function to generate $3$ different signal whose modulation values are $0.5, 1, 2.$ 

```python
sig_1 = modulate(message, A_c, f_c, 0.5)
sig_2 = modulate(message, A_c, f_c, 1)
sig_3 = modulate(message, A_c, f_c, 2)
```

### Plots

It is time to plot the signals. You can see the how different values of modulation index ($\mu$) affects the modulation of message signal. For $\mu = 0.5$, we say signal is under-modulated. And for $\mu = 2$, we say it is over-modulated.  

```python
plt.rcParams['figure.figsize'] = [14, 7]

plt.subplot(2,2,1)
plt.plot(t, message, 'b')
plt.title("Message Signal")
plt.xlabel('$t$ (sec)')
plt.ylabel('$m(t)$')

plt.subplot(2,2,2)
plt.plot(t, sig_1, 'g')
plt.title("Modulated Signal ( $\mu = 0.5$ )")
plt.xlabel('$t$ (sec)')
plt.ylabel('$s_{1}(t)$')

plt.subplot(2,2,3)
plt.plot(t, sig_2, 'r')
plt.title("Modulated Signal ( $\mu = 1$ )")
plt.xlabel('$t$ (sec)')
plt.ylabel('$s_{2}(t)$')

plt.subplot(2,2,4)
plt.plot(t, sig_3, 'orange')
plt.title("Modulated Signal ( $\mu = 2$ )")
plt.xlabel('$t$ (sec)')
plt.ylabel('$s_{3}(t)$')

plt.tight_layout()
plt.show()
```


### Creating Frequency Spectrum

```python
plt.subplot(3,1,1)
ft = np.fft.fft(sig_1)
ft = np.fft.fftshift(ft)
freq = np.fft.fftfreq(len(ft))
plt.plot(freq, np.absolute(ft), 'g')
plt.title("Modulated Signal ( $\mu = 0.5$ )")
plt.xlabel('$f$ (Hz)')
plt.ylabel('$S_{1}(f)$')

plt.subplot(3,1,2)
ft = np.fft.fft(sig_2)
ft = np.fft.fftshift(ft)
freq = np.fft.fftfreq(len(ft))
plt.plot(freq, np.absolute(ft), 'r')
plt.title("Modulated Signal ( $\mu = 1$ )")
plt.xlabel('$f$ (Hz)')
plt.ylabel('$S_{2}(f)$')

plt.subplot(3,1,3)
ft = np.fft.fft(sig_3)
ft = np.fft.fftshift(ft)
freq = np.fft.fftfreq(len(ft))
plt.plot(freq, np.absolute(ft), 'orange')
plt.title("Modulated Signal ( $\mu = 2$ )")
plt.xlabel('$f$ (Hz)')
plt.ylabel('$S_{3}(f)$')

plt.tight_layout()
plt.show()
```
